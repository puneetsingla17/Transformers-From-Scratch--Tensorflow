```python
def positional_embedding(pos, model_size):
    PE = np.zeros((1, model_size))
    for i in range(model_size):
        if i % 2 == 0:
            PE[:,i] = np.sin(pos / 10000 ** (i / model_size))
        else:
            PE[:,i] = np.cos(pos / 10000 ** ((i - 1) / model_size))
    return PE

# max_length = max(len(data_en[0]), len(data_fr_in[0]))
MODEL_SIZE = 128

pes = []
for i in range(maxenglen):
    pes.append(positional_embedding(i, MODEL_SIZE))

pes = np.concatenate(pes, axis=0)
pes = tf.constant(pes, dtype=tf.float32)
```

```python
def multiheadattention(modelsize,h):
    querysize=modelsize//h
    keysize=modelsize//h
    valuesize=modelsize//h
    wq=[tf.keras.layers.Dense(querysize) for _ in range(h)]
    wk=[tf.keras.layers.Dense(keysize) for _ in range(h)]
    wv=[tf.keras.layers.Dense(valuesize) for _ in range(h)]
    wo=tf.keras.layers.Dense(modelsize)
    
    return wq,wk,wv,wo,keysize
    
```

```python
def call(query,value,keysize,h,wq,wk,wv,wo):
    # query has shape (batch , querylen , modelsize)
    # value has shape (batch , valuelen , modelsize)
    heads=[]
    for i in range(h):
        score=tf.matmul(wq[i](query),wk[i](value),transpose_b=True)
        
        score/=tf.math.sqrt(tf.dtypes.cast(keysize,tf.float32))
        
        alignment=tf.nn.softmax(score,axis=2)
        
        head=tf.matmul(alignment,wv[i](value))    # head has shape  batch x decoderlen x valuesize
        
        heads.append(head)
        
    heads=tf.concat(heads,axis=2)
    heads=wo(heads)      # heards has shape batch x querylen x modelsize
    
    return heads
```
